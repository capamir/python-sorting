```python
import matplotlib.pyplot as plt

# Given data
list_length = [1000, 5000, 10000, 50000, 100000, 200000, 300000]
bubble_sort = [0.0, 0.145, 0.6, 14.05, 56.9, 229.44, 513.34]
insertion_sort = [0.0, 0.045, 0.18, 4.25, 17.62, 69.12, 154.22]
quick_sort = [0.0, 0.0, 0.0, 0.01, 0.02, 0.05, 0.065]
```

```python
# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(list_length, bubble_sort, label='Bubble Sort', marker='o')
plt.plot(list_length, insertion_sort, label='Insertion Sort', marker='s')
plt.plot(list_length, quick_sort, label='Quicksort', marker='^')

# Add labels and title
plt.xlabel('List Length')
plt.ylabel('Runtime (seconds)')
plt.title('Sorting Algorithms Runtime vs. List Length')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
